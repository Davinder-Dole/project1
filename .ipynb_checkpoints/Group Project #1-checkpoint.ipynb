```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
df = pd.read_csv("data/Crime_Data_from_2010_to_Present.csv")
df.columns
```

```python
df["Date Occurred"] = pd.to_datetime(df["Date Occurred"])
df["Date Reported"] = pd.to_datetime(df["Date Reported"])
df.to_csv("crime_data_1.csv")
```

```python
df.dtypes
```

```python
data_2010 = df.loc[(df["Date Occurred"] < datetime(2011, 1, 1)) & (df["Date Occurred"] >= datetime(2010, 1, 1))] 
data_2010.head()
data_2010.shape
data_2010.to_csv("data_2010.csv")
```

```python
data_2017 = df.loc[(df["Date Occurred"] < datetime(2018, 1, 1)) & (df["Date Occurred"] >= datetime(2017, 1, 1))]
data_2017.to_csv("data_2017.csv")
data_2017.shape
```

```python
data_2017.head(100)
```

```python
a = data_2017.loc[~data_2017["Location "].isna()]
a['Location ']
```

```python
data_2017.count()
```

```python
area_info = data_2017.groupby("Area Name")["Reporting District"].count()
area_info.reset_index

area_info = area_info.sort_values(ascending=True)
print(area_info)
```

```python
plt.bar(range(len(area_info)), area_info)
plt.title("Crime by Area in 2017")
plt.xlabel("")
plt.ylabel("")
tick_locations = [value for value in np.arange(len(area_info))]
labels = [values for values in area_info.index]
plt.xticks(tick_locations, labels, rotation="vertical")
```
